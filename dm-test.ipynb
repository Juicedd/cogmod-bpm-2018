# Declarative memory test

```python
from dm import DM # import from dm.py in the same directory
import numpy as np
import matplotlib.pyplot as plt
```

Create a new DM:

```python
dm = DM()
```

Add some encounters of a few different chunks. They can be any type (e.g. number of pulses) but for clarity I'm using strings here. They are stored as key-value pairs in a dictionary. The key is the content of the chunk, the value is a list of encounters.

```python
dm.add_encounter(content = "banana", time = 0)
dm.add_encounter("orange", 2)
dm.add_encounter("pineapple", 6)
dm.add_encounter("mango", 0)
dm.add_encounter("mango", 1.5)
dm.add_encounter("mango", 3.4)
dm.add_encounter("mango", 6.99)
dm.add_encounter("mango", 10.0)

print(dm)
```

What is the baselevel activation of the "mango" chunk over time?

```python
x = np.linspace(start = 0, stop = 15, num = 1000)
bl = []
for i in x:
    bl.append(dm.get_baselevel_activation("mango", i + 0.001))
    
plt.plot(x, bl)
```

What is the retrieval latency of the "mango" chunk over time?

```python
x = np.linspace(0, 15, 1000)
lat = []
for i in x:
    lat.append(dm.get_latency("mango", i + 0.001))
    
plt.plot(x, lat)
```
